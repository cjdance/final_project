```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
bout_sum_csv = "bout_summary.csv"
fighter_csv = "fighter_data.csv"
```

```python
bout_df = pd.read_csv(bout_sum_csv)
bout_df.head()
```

```python
## List of transformations to be performed

# remove fighter id columns from the dataset
# drop any fights with an unlimited time limit
# drop location data, it will not be used in this project

bout_df = bout_df[["fight_id","fighter_1", "fighter_2", "winner", "decision_method", "fight_duration_lastrnd", "fight_duration_lastrnd_time",
                 "time_format", "weight_class", "date"]]
```

```python
bout_df["time_format"].unique()
```

```python
bout_df = bout_df.drop(bout_df.index[bout_df["time_format"] != '3 Rnd (5-5-5)'])
bout_df
```

```python
new = bout_df["fight_duration_lastrnd_time"].str.split(":")    
```

```python
bout_df["fight_max_length"] = dt.timedelta(minutes=5) * bout_df["fight_duration_lastrnd"]
```

```python
bout_end_times = []

for i in new:
    end_time = dt.timedelta(minutes=int(i[0]), seconds=int(i[1]))
    bout_end_times.append(end_time)
    
bout_df["bout_end_time"] = bout_end_times
```

```python
fight_length = []

for i in bout_df["bout_end_time"]:
    j = 0
    if i == dt.timedelta(minutes=5):
        bout_length = bout_df["fight_max_length"][j].total_seconds()/60
        j = j + 1
        
    else:
        bout_length = bout_df["fight_max_length"][j] - i
        bout_length = bout_length.total_seconds()/60
        j = j + 1
    
    fight_length.append(bout_length)
    
bout_df["fight_length_min"] = fight_length
```

```python
bout_df = bout_df.drop("fight_max_length", axis = 1)
bout_df = bout_df.drop("bout_end_time", axis=1)
```

```python
new = bout_df["date"].str.split(",")    
years = []

for i in new:
    year = i[1]
    years.append(year)
    
bout_df["date"] = years
```

```python
bout_df = bout_df.drop(["fighter_1", "fighter_2"], axis = 1)
bout_df
```

```python
fighter_df = pd.read_csv(fighter_csv)
fighter_df
```

```python
## List of transformations to be performed

# remove fighter_id column, we need to use the index as the fighter id when dropping into postgresSql
# convert DOB column into datetime format for ease of use in later calculations
# create a total fights column which adds draws/losses/wins for a total
# check for duplicate fighter names and drop the fighter with the lowest number of total fights (more fights means more data for their stats)
# drop any fighter with null values or zeroes in the sig_str_abs_PM and other columns, zeroes or nulls here would indicate they have no fight data
# add a weightclass column alongside fighter weight
# check for null values in stance column and replace with "Orthodox", fighters with no stance will be assumed for now (i.e. right handed)
```

```python
fighter_df = fighter_df.drop("fighter_id", axis=1)
```

```python
fighter_df["total_fights"] = fighter_df["n_draw"] + fighter_df["n_loss"] + fighter_df["n_win"]
```

```python
fighter_df = fighter_df.dropna(subset=['sig_str_abs_pM', 'sig_str_def_pct', 'sig_str_land_pM', 'sig_str_land_pct','dob'])
```

```python
fighter_df = fighter_df.drop((fighter_df.index[fighter_df["dob"] == '--']))
```

```python
fighter_df = fighter_df.drop((fighter_df.index[fighter_df["sig_str_abs_pM"] == 0]))
```

```python
fighter_df = fighter_df.reset_index(drop=True)
```

```python
new = fighter_df["dob"].str.split(",")    
years = []

for i in new:
    year = i[1]
    years.append(year)
    
fighter_df['dob'] = years
```

```python
fighter_df
```

```python
fighter_df["dob"] = fighter_df["dob"].astype('int')
new = fighter_df["reach"].str.split('"')
reach = []
for i in new:
    x = i[0]
    reach.append(x)
fighter_df["reach"] = reach

fighter_df = fighter_df.replace({"stance": "--", "stance": "switch", "stance": "Sideways"}, "Orthodox")
fighter_df = fighter_df.replace({"stance": "Open Stance"}, "Southpaw")
fighter_df = fighter_df.replace({"stance": "Switch"}, "Orthodox")
values = {"stance": "Orthodox"}
fighter_df = fighter_df.fillna(value = values)

fighter_df = fighter_df.replace({"height": "--"}, "")
fighter_df = fighter_df.replace({"reach": "--"}, "")
```

```python
bout_detail_df = pd.read_json('bout_data.jl', lines =True)
bout_detail_df
```

```python
## Transformations required

# identify unneccessary columns i.e remove
# 'fighter_id', 'n_pass', 'n_rev', 'total_strike_def', 'td_abs', 'td_def', 'head_abs', 'head_att', 'head_def','head_abs',
#       'head_att', 'head_def', 'head_land', 'body_abs', 'body_att', 'body_def',
#      'body_land', 'leg_abs', 'leg_att', 'leg_def', 'leg_land',
#       'distance_abs', 'distance_att', 'distance_def', 'distance_land',
#       'clinch_abs', 'clinch_att', 'clinch_def', 'clinch_land', 'ground_abs',
#       'ground_att', 'ground_def', 'ground_land' 
# spearate doubled up columns and break into fighter_1, fighter_2 columns for each, create a new dataframe for this to preserve raw data
# after new dataframe is created, further manipulation is required after joining the dataframe with bout duration
# convert remaining striking columns into averages and percentages over the duration of the bout to align with fighter stats
```

```python
new = bout_detail_df["fighter_name"]

for i in new:
    
    fighter_1_nm = i[0]
    fighter_2_nm = i[1]
    
    fighter_1.append(fighter_1_nm)
    fighter_2.append(fighter_2_nm)
    

    
for i in bout_detail_df["kd"]:
    
    i = bout_detail_df["kd"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_kd = int(i[0])
    f2_kd = int(i[1])
    
    fighter_1_kd.append(f1_kd)
    fighter_2_kd.append(f2_kd)
    
for i in bout_detail_df["n_sub"]:
    
    i = bout_detail_df["n_sub"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")

    f1_nsub = int(i[0])
    f2_nsub = int(i[1])
    
    fighter_1_nsub.append(f1_nsub)
    fighter_2_nsub.append(f2_nsub)
    
for i in bout_detail_df["sig_str_abs"]:
    
    i = bout_detail_df["sig_str_abs"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_sig_abs = int(i[0])
    f2_sig_abs = int(i[1])
    
    fighter_1_sig_str_abs.append(f1_sig_abs)
    fighter_2_sig_str_abs.append(f2_sig_abs)
    
for i in bout_detail_df["sig_str_att"]:
    
    i = bout_detail_df["sig_str_att"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_sig_att = int(i[0])
    f2_sig_att = int(i[1])
    
    fighter_1_sig_str_att.append(f1_sig_att)
    fighter_2_sig_str_att.append(f2_sig_att)
    
for i in bout_detail_df["sig_str_def"]:
    
    i = bout_detail_df["sig_str_def"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_sig_def = int(i[0])
    f2_sig_def = int(i[1])
    
    fighter_1_sig_str_def.append(f1_sig_def)
    fighter_2_sig_str_def.append(f2_sig_def)
    
for i in bout_detail_df["sig_str_land"]:
    
    i = bout_detail_df["sig_str_land"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_sig_land = int(i[0])
    f2_sig_land = int(i[1])
    
    fighter_1_sig_str_land.append(f1_sig_land)
    fighter_2_sig_str_land.append(f2_sig_land)
    
for i in bout_detail_df["total_str_abs"]:
    
    i = bout_detail_df["total_str_abs"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_tot_abs = int(i[0])
    f2_tot_abs = int(i[1])
    
    fighter_1_tot_str_abs.append(f1_tot_abs)
    fighter_2_tot_str_abs.append(f2_tot_abs)
    
for i in bout_detail_df["total_str_att"]:
    
    i = bout_detail_df["total_str_att"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_tot_att = int(i[0])
    f2_tot_att = int(i[1])
    
    fighter_1_tot_str_att.append(f1_tot_att)
    fighter_2_tot_str_att.append(f2_tot_att)
    
for i in bout_detail_df["total_str_def"]:
    
    i = bout_detail_df["total_str_def"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_tot_def = int(i[0])
    f2_tot_def = int(i[1])
    
    fighter_1_tot_str_def.append(f1_tot_def)
    fighter_2_tot_str_def.append(f2_tot_def)
    
for i in bout_detail_df["total_str_land"]:
    
    i = bout_detail_df["total_str_land"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_tot_land = int(i[0])
    f2_tot_land = int(i[1])
    
    fighter_1_tot_str_land.append(f1_tot_land)
    fighter_2_tot_str_land.append(f2_tot_land)
    
for i in bout_detail_df["td_att"]:
    
    i = bout_detail_df["td_att"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_td_att = int(i[0])
    f2_td_att = int(i[1])
    
    fighter_1_td_att.append(f1_td_att)
    fighter_2_td_att.append(f2_td_att)
    
for i in bout_detail_df["td_land"]:
    
    i = bout_detail_df["td_land"].astype('string')

    i = i[0].split("[")
    i = i[1].split("]")
    i = i[0].split(",")
    
    f1_td_land = int(i[0])
    f2_td_land = int(i[1])
    
    fighter_1_td_land.append(f1_td_land)
    fighter_2_td_land.append(f2_td_land)
    
```
