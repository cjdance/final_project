```python
## List of transformations to be performed

# remove fighter id columns from the dataset
# drop any fights with an unlimited time limit
# drop location data, it will not be used in this project

bout_df = bout_df[["fight_id","fighter_1", "fighter_2", "winner", "decision_method", "fight_duration_lastrnd", "fight_duration_lastrnd_time",
                 "time_format", "weight_class", "date"]]
```

```python
new = bout_df["fight_duration_lastrnd_time"].str.split(":")    
```

```python
bout_df["fight_max_length"] = dt.timedelta(minutes=5) * bout_df["fight_duration_lastrnd"]
```

```python
bout_end_times = []

for i in new:
    end_time = dt.timedelta(minutes=int(i[0]), seconds=int(i[1]))
    bout_end_times.append(end_time)
    
bout_df["bout_end_time"] = bout_end_times
```

```python
fight_length = []

for i in bout_df["bout_end_time"]:
    j = 0
    if i == dt.timedelta(minutes=5):
        bout_length = bout_df["fight_max_length"][j].total_seconds()/60
        j = j + 1
        
    else:
        bout_length = bout_df["fight_max_length"][j] - i
        bout_length = bout_length.total_seconds()/60
        j = j + 1
    
    fight_length.append(bout_length)
    
bout_df["fight_length_min"] = fight_length
```

```python
bout_df = bout_df.drop("fight_max_length", axis = 1)
bout_df = bout_df.drop("bout_end_time", axis=1)
```

```python
## Data pre-processing for the model
```

```python
X = data_df.drop("outcome", axis=1)
y = data_df["outcome"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)

X_test_scaled = X_scaler.transform(X_test)
```

```python
label_encoder = LabelEncoder()

label_encoder.fit(y_train)

encoded_y_train = label_encoder.transform(y_train)

encoded_y_test = label_encoder.transform(y_test)
```

```python
y_train_categorical = to_categorical(encoded_y_train)

y_test_categorical = to_categorical(encoded_y_test)
```

```python
# Create Deep-Learning Model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(units=100, activation='relu', input_dim=14))

model.add(Dense(units=100, activation='relu'))

model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)

print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes()

prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```
