```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
bout_sum_csv = "bout_summary.csv"
fighter_csv = "fighter_data.csv"
```

```python
bout_df = pd.read_csv(bout_sum_csv)
bout_df.head()
```

```python
## List of transformations to be performed

# remove fighter id columns from the dataset
# drop any fights with an unlimited time limit
# drop location data, it will not be used in this project

bout_df = bout_df[["fight_id","fighter_1", "fighter_2", "winner", "decision_method", "fight_duration_lastrnd", "fight_duration_lastrnd_time",
                 "time_format", "weight_class", "date"]]
```

```python
bout_df["time_format"].unique()
```

```python
bout_df = bout_df.drop(bout_df.index[bout_df["time_format"] != '3 Rnd (5-5-5)'])
bout_df
```

```python
new = bout_df["fight_duration_lastrnd_time"].str.split(":")    
```

```python
bout_df["fight_max_length"] = dt.timedelta(minutes=5) * bout_df["fight_duration_lastrnd"]
```

```python
bout_end_times = []

for i in new:
    end_time = dt.timedelta(minutes=int(i[0]), seconds=int(i[1]))
    bout_end_times.append(end_time)
    
bout_df["bout_end_time"] = bout_end_times
```

```python
fight_length = []

for i in bout_df["bout_end_time"]:
    j = 0
    if i == dt.timedelta(minutes=5):
        bout_length = bout_df["fight_max_length"][j].total_seconds()/60
        j = j + 1
        
    else:
        bout_length = bout_df["fight_max_length"][j] - i
        bout_length = bout_length.total_seconds()/60
        j = j + 1
    
    fight_length.append(bout_length)
    
bout_df["fight_length_min"] = fight_length
```

```python
bout_df = bout_df.drop("fight_max_length", axis = 1)
bout_df = bout_df.drop("bout_end_time", axis=1)
```

```python
new = bout_df["date"].str.split(",")    
years = []

for i in new:
    year = i[1]
    years.append(year)
    
bout_df["date"] = years
```

```python
fighter_df = fighter_df.reset_index(drop=True)
```

```python
new = fighter_df["dob"].str.split(",")    
years = []

for i in new:
    year = i[1]
    years.append(year)
    
fighter_df['dob'] = years
```
